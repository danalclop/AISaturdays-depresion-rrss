```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

**Read data into a Pandas DataFrame**

```python
data = pd.read_csv('OSF_socialmedia_data.csv')
data.head()
```

```python
data.shape
```

```python
nullcol = data.columns[data.isna().any()]
nullsum = data[nullcol].isna().sum()
responded = data['Responded'].value_counts()
print('{} sessions answered, {} left without answer \n'.format(responded[1],responded[0]))
print(nullsum)
print("\n Dentro de las sesiones que sí se han contestado, hay valores nulos")
```

```python
#De este modo tendremos un dataframe sin las sesiones sin contestar
answered = data.loc[:, 'Responded'] == 1
cleaned_data = data.loc[answered]

#Pasemos a ver cuantos valores nulos tenemos dentro de las sesiones contestadas
nullcol = cleaned_data.columns[cleaned_data.isna().any()]
nullsum = cleaned_data[nullcol].isna().sum()
print('Nulos en el dataset limpio \n {}'.format(nullsum))

#Vamos a ver los valores del dataset
print('\n Valores del dataset limpio \n')
surveydata = cleaned_data.iloc[:,14:]
surveycolumns = surveydata.columns
for col in surveycolumns:
    print(cleaned_data[col].value_counts())
    print(cleaned_data[col].value_counts(normalize=True)[0])
    print()
#¿Muchos ceros?¿Normal?
```
