```python
import pandas as pd
```

```python
data = pd.read_csv('../raw-data/EDA-OSF_socialmedia_data.csv', index_col=0)
```

```python
data.drop('Unnamed: 0.1', axis=1, inplace=True)
```

```python
data.info()
```

```python
data['Date']= pd.to_datetime(data['Date'])
```

```python
# data['Survived'][data['Sex'] == 'male']).mean()

for participant in data.Participant.unique():
    print("Paricipant ID is: {}".format(participant))
    for date in data['Date'][data['Participant'] == participant].unique():
        print("This participant was measured on: {}".format(date))
        print("Averate LowMood score on this day was {:.2f}".format(data['DeprMood'][data['Date'] == date].mean()))
```
